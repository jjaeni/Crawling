### 인스타그램 태그 검색어로 이미지 크롤링

검색했을때 나오는 약 30장만 크롤링 가능

```python
from urllib.request import urlopen
from urllib.parse import quote_plus as qp #아스키코드로 변환
from bs4 import BeautifulSoup
from selenium import webdriver
import time



# 검색어 입력
hastag = input('검색할 해시태그 입력: ')
url = 'https://www.instagram.com/explore/tags/' + qp(hastag)
print(url)

# 공통 클래스 > 내용 추출
driver = webdriver.Chrome()
driver.get(url)
time.sleep(3) #딜레이 방지용
html = driver.page_source
soup = BeautifulSoup(html)

insta = soup.select('._aabd._aa8k._aanf')

n = 1
for i in insta:
    print("https://www.instagram.com"+i.a['href']) # a태그의 속성 href만 가져오기
    imgUrl = i.select_one('._aagv').img['src']
    
    with urlopen(imgUrl) as f:
        with open('./img/' + hastag + str(n) + '.jpg','wb') as h:
            img = f.read()
            h.write(img)
        n += 1
        
driver.close()
```

### 인스타그램 검색어로 이미지 크롤링 2

스크롤 내리고, 페이지 바꿔가며 이미지 저장 가능하도록

```python
from urllib.request import urlopen
from urllib.parse import quote_plus as qp #아스키코드로 변환
from bs4 import BeautifulSoup
from selenium import webdriver
import time, requests, shutil


hastag = input('검색할 해시태그 입력: ')
url = 'https://www.instagram.com/explore/tags/' + qp(hastag)
print(url)

driver = webdriver.Chrome()
driver.get(url)

time.sleep(3)
ß
html = driver.page_source
soup = BeautifulSoup(html)

imglist = []

for j in range(0, 15): #스크롤 내리는 반복 횟수
    insta = soup.select('._aabd._aa8k._aanf')
    
    for i in insta:
        print("https://www.instagram.com"+i.a['href']) # a태그의 속성 href만 가져오기
        imgUrl = i.select_one('._aagv').img['src']
        imglist.append(imgUrl)
        imglist = list(set(imglist))
        html = driver.page_source
        soup = BeautifulSoup(html)
        insta = soup.select('._aabd._aa8k._aanf')
    
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(2)
    
n = 0

for i in range(0,100):
    # this is the image url
    image_url = imglist[n]
    # open the url image, set stream to True, this will return the stream content.

    resp = requests.get(image_url, stream=True)
    #Open a local file with wb{write binary} permission

    local_file = open('./img/' + hastag + str(n) + '.jpg','wb')
    # Set decode_content value to True, otherwise the downloaded image file's size will be zero
    resp.raw.decode_content = True
    # Copy the response stream raw data to local image file
    shutil.copyfileobj(resp.raw, local_file)
    # Remove the image url response object
    n += 1
    del resp
    
        
driver.close()
```
